```python
import pandas as pd

df = pd.read_csv(r'C:\Users\venu krishna\Downloads\cricket.csv') 

df
```

```python
df.info()
```

```python
d1 = {'Sunny':0, 'Overcast':1, 'Rain':2}

df['Outlook'] = df['Outlook'].map(d1)

df
```

```python
d2 = {'Hot':0, 'Mild':1, 'Cool':2}

df['Temperature'] = df['Temperature'].map(d2)

df
```

```python
d3 = {'High':0, 'Normal':1}

df['Humidity'] = df['Humidity'].map(d3)

df
```

```python
d4 = {'Weak':0, 'Strong':1}

df['Wind'] = df['Wind'].map(d4)

df
```

```python
d5 = {'No':0, 'Yes':1}

df['Decision'] = df['Decision'].map(d5)

df
```

```python
import pandas as pd

from sklearn import tree

from sklearn.tree import DecisionTreeClassifier

import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder

from sklearn import metrics
```

```python
X1 = ['Outlook', 'Temperature', 'Humidity', 'Wind']

X = df[X1] 

y = df['Decision']

model1 = tree.DecisionTreeClassifier()

model1 = model1.fit(X, y)
```

```python
tree.plot_tree(model1)
```

```python
y_predict = model1.predict(X)

from sklearn.metrics import accuracy_score

accuracy_score(y, y_predict)
```

```python
from sklearn.metrics import confusion_matrix

pd.DataFrame(confusion_matrix(y, y_predict),
    columns=['Predicted No match', 'Predicted match'],
    index=['Not supported ', 'supported weather'])
```
